```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('heart_2020_cleaned.csv')
df
```

```python
#Histogram distribution demonstrating the Mental Health "ratings" for Males
df['MentalHealth'].where(df['Sex'] == 'Male' ).plot(kind='hist')
```

```python
#Histogram distribution demonstrating the Mental Health "ratings" for Males
df['MentalHealth'].where(df['Sex'] == 'Female' ).plot(kind='hist')
```

```python
#Plots Frequency vs. Mental Health for Smoking Males on one graph and combines it with Frequency vs. Mental Health for Smoking Females of another graph
legends = ["BMI", "Physcial Health", "Mental Health"]
df[legends].where(df['Sex'] == 'Male').mean().plot(kind = "line")
```

How does being a male or female effect certain Mental Health conditions given constant negative impacts (inlcuding but not limitied to Smoking, Drinking, Strokes, Bad Physical Health/BMI, and more)?
